```python
import numpy as np
import pandas as pd
```

```python
train_data = pd.read_csv('D:/Data Science/Course_Quera/Class II/Project1/canceli/data/train_data.csv')
train_data.head()
```

```python
train_data.info()
```

```python
test_data = pd.read_csv('D:/Data Science/Course_Quera/Class II/Project1/canceli/data/test_data.csv')
test_data.head()
```

```python
test_data.info()
```

```python
from datetime import datetime as dt
from datetime import date
from datetime import timedelta

train_data['Duration'] = pd.to_datetime(train_data.DepartureTime) - pd.to_datetime(train_data.Created)
train_data['Duration'] = train_data['Duration'].dt.days

test_data['Duration'] = pd.to_datetime(test_data.DepartureTime) - pd.to_datetime(test_data.Created)
test_data['Duration'] = test_data['Duration'].dt.days
```

```python
train_data['TicketsPerOrder'] = train_data.groupby('BillID')['TicketID'].transform('count')
test_data['TicketsPerOrder'] = test_data.groupby('BillID')['TicketID'].transform('count')
```

```python
train_data['is_with_family'] = train_data.groupby('BillID')['Male'].transform('count')
test_data['is_with_family'] = test_data.groupby('BillID')['Male'].transform('count')
```

```python
from sklearn.preprocessing import LabelEncoder 
le = LabelEncoder() 

train_data['TripReason'] = le.fit_transform(train_data['TripReason'])
test_data['TripReason'] = le.transform(test_data['TripReason'])

train_data['Vehicle'] = le.fit_transform(train_data['Vehicle'])
test_data['Vehicle'] = le.transform(test_data['Vehicle'])

# combine both columns From and To in train and test sets as a new set, since they may be different 
Cities = set(train_data['From']) | set(train_data['To']) | set(test_data['From']) | set(test_data['To'])
city = pd.DataFrame(Cities, columns=['City'])
city['City'] = le.fit_transform(city['City'])

train_data['From'] = le.transform(train_data['From'])
test_data['From'] = le.transform(test_data['From'])

train_data['To'] = le.transform(train_data['To'])
test_data['To'] = le.transform(test_data['To'])
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

##The input to MinMaxScaler needs to be array-like, with shape [n_samples, n_features].
##So you can apply it on the column as a dataframe rather than a series (using double square brackets instead of single)

train_data['Price'] = scaler.fit_transform(train_data[['Price']])
test_data['Price'] = scaler.transform(test_data[['Price']])
```

```python
MyList = ['Created', 'DepartureTime', 'UserID', 'VehicleType', 'VehicleClass', 'HashPassportNumber_p', 'HashEmail', 'BuyerMobile', 'NationalCode']
train_data.drop(columns=MyList, axis=1, inplace=True)
test_data.drop(columns=MyList, axis=1, inplace=True)
```

```python
y = train_data.Cancel
X = train_data.drop(['CancelTime', 'Cancel'], axis=1)

from sklearn.model_selection import train_test_split
X_train, X_valid, y_train, y_valid = train_test_split(X,y , test_size=0.2)
```

```python
# modeling 
from xgboost import XGBClassifier

# create model instance
bst = XGBClassifier(n_estimators=4, max_depth=3)
# fit model
bst.fit(X_train, y_train)
# make predictions
y_pred = bst.predict(X_valid)

# evaluate model
from sklearn.metrics import  f1_score, classification_report, confusion_matrix
f1_score(y_valid, y_pred, average='weighted')
```

```python
# predict test samples
prediction = bst.predict(test_data)
submission = pd.DataFrame(prediction, columns=['Cancel'])
submission
```
